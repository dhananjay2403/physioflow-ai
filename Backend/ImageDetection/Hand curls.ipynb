```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

def calculate_angle(a, b, c):
    a = np.array(a) # First point
    b = np.array(b) # Mid point
    c = np.array(c) # End point
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle > 180.0:
        angle = 360-angle
        
    return angle

# Function to create rounded rectangle
def draw_rounded_rectangle(img, top_left, bottom_right, radius, color, thickness=-1):
    # Draw the main rectangle, leaving the corners
    x1, y1 = top_left
    x2, y2 = bottom_right
    
    # Draw the filled rectangle without corners
    cv2.rectangle(img, (x1 + radius, y1), (x2 - radius, y1 + radius), color, thickness)  # Top
    cv2.rectangle(img, (x1, y1 + radius), (x2, y2 - radius), color, thickness)  # Middle
    cv2.rectangle(img, (x1 + radius, y2 - radius), (x2 - radius, y2), color, thickness)  # Bottom
    
    # Draw the four corner circles
    cv2.circle(img, (x1 + radius, y1 + radius), radius, color, thickness)  # Top-left
    cv2.circle(img, (x2 - radius, y1 + radius), radius, color, thickness)  # Top-right
    cv2.circle(img, (x1 + radius, y2 - radius), radius, color, thickness)  # Bottom-left
    cv2.circle(img, (x2 - radius, y2 - radius), radius, color, thickness)  # Bottom-right

cap = cv2.VideoCapture(0)

# Curl counter variables for both arms
left_counter = 0 
left_stage = None
right_counter = 0
right_stage = None

# UI parameters
panel_width = 400
panel_height = 100
corner_radius = 25  # 25% of height would be 25 pixels for a 100px height panel
bg_color = (255, 255, 255)  # White background
text_color = (75, 75, 75)   # Dark gray text
accent_color = (0, 128, 255)  # Blue accent

## Setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
      
        # Make detection
        results = pose.process(image)
    
        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            
            # Get coordinates for LEFT arm
            left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            
            # Get coordinates for RIGHT arm
            right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
            
            # Calculate angles
            left_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
            right_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
            
            # Visualize angles with more modern styling
            for side, elbow, angle in [("L", left_elbow, left_angle), ("R", right_elbow, right_angle)]:
                # Get coordinates and adjust
                coord = tuple(np.multiply(elbow, [640, 480]).astype(int))
                
                # Draw a semi-transparent background circle
                cv2.circle(image, coord, 30, (0, 0, 0, 0.3), -1)
                
                # Add angle text
                cv2.putText(image, f"{int(angle)}°", 
                           (coord[0]-20, coord[1]+5), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2, cv2.LINE_AA)
            
            # Curl counter logic for left arm
            if left_angle > 160:
                left_stage = "down"
            if left_angle < 30 and left_stage == 'down':
                left_stage = "up"
                left_counter += 1
            
            # Curl counter logic for right arm
            if right_angle > 160:
                right_stage = "down"
            if right_angle < 30 and right_stage == 'down':
                right_stage = "up"
                right_counter += 1
                       
        except:
            pass
        
        # Create modern UI panel
        # First create a clean image for our UI panel to avoid transparency issues
        h, w, _ = image.shape
        panel_x = (w - panel_width) // 2  # Center the panel
        panel_y = 20  # Top margin
        
        # Draw the main rounded panel
        draw_rounded_rectangle(
            image, 
            (panel_x, panel_y), 
            (panel_x + panel_width, panel_y + panel_height), 
            corner_radius, 
            bg_color, 
            -1
        )
        
        # Draw divider
        divider_x = panel_x + panel_width // 2
        cv2.line(image, 
                (divider_x, panel_y + 15), 
                (divider_x, panel_y + panel_height - 15), 
                (200, 200, 200), 2, cv2.LINE_AA)
        
        # Left arm section
        cv2.putText(image, 'LEFT ARM', (panel_x + 20, panel_y + 25), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, text_color, 1, cv2.LINE_AA)
        
        # Left counter with accent background
        counter_radius = 15
        counter_x = panel_x + 40
        counter_y = panel_y + 60
        cv2.circle(image, (counter_x, counter_y), counter_radius, accent_color, -1)
        cv2.putText(image, str(left_counter), 
                   (counter_x - 7 if left_counter < 10 else counter_x - 13, counter_y + 7), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)
        
        # Left stage
        stage_text = left_stage.upper() if left_stage else "READY"
        cv2.putText(image, stage_text, 
                   (panel_x + 70, panel_y + 65), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, text_color, 2, cv2.LINE_AA)
        
        # Right arm section
        cv2.putText(image, 'RIGHT ARM', (divider_x + 20, panel_y + 25), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, text_color, 1, cv2.LINE_AA)
        
        # Right counter with accent background
        counter_x = divider_x + 40
        cv2.circle(image, (counter_x, counter_y), counter_radius, accent_color, -1)
        cv2.putText(image, str(right_counter), 
                   (counter_x - 7 if right_counter < 10 else counter_x - 13, counter_y + 7), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)
        
        # Right stage
        stage_text = right_stage.upper() if right_stage else "READY"
        cv2.putText(image, stage_text, 
                   (divider_x + 70, panel_y + 65), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, text_color, 2, cv2.LINE_AA)
        
        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2), 
                                mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2) 
                                )
        
        cv2.imshow('Bicep Curl Counter', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()
```
